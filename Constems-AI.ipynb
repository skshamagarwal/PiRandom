**Part 1: Finding Random Numbers Without using any internal or external Python Libraries**

```python
# Using Linear Feedback Shift Register
def rand(n, x, y):
    state = (1 << 127) | 1
    ele = []
    for i in range(n):
        bit = (state ^ (state >> 2) ^ (state >> 3 ) ^ (state >> 5))&1
        state = (state >> 1) | (bit << 127)
        res = state%y
        if res<x:
            ele.append(res+x)
        else: 
            ele.append(state%y)
    return ele
```

```python
n = int(input("Enter no. of random elements you want to generate :: "))
x, y = map(int, input("Enter Range :: ").split())
var = rand(n, x, y)
print()
print(var)
```

**Part 3: Predicting the value of pi using Random Numbers**

```python
# Finding GCD
def gcd(a, b):
    if b>a:
        b,a = a,b
    r = a%b
    if r==0:
        return b
    else:
        return gcd(b, r)
```

```python
import math

coPrimeCount = 0
total = 0
for i in range(0, len(var)-1, 2):
    var_a = var[i]
    var_b = var[i+1]
    if (gcd(var_a,var_b)==1):
        coPrimeCount+=1
    total+=1


percentCoprime = (total-coPrimeCount) / total;
# percentCoprime = coPrimeCount / total;
pie = math.sqrt((6.0/percentCoprime))
print(pie)
```
